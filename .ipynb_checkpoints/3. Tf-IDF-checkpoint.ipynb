# Text analysis of titles using sklearn Tf-IDF

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
data=pd.read_csv('data.csv')
titles=np.array(data['title'])
```

```python
#removing the digits
from string import digits
remove_digits=str.maketrans('','',digits)
titles=[x.translate(remove_digits) for x in titles]
titles=[i for i in titles if len(i.split(' '))>1]
```

```python
u=TfidfVectorizer(min_df=5,max_df=0.5)
weights=u.fit_transform(titles)
t=weights.toarray()
```

```python
def get_closest_dist(mat,ind):
    from numpy.linalg import norm
    current_min=100000
    min_ind=0
    print(titles[ind])
    for i in range(len(mat)):
        if not i==ind:
            if np.dot( mat[i] , mat[ind] )!=0:
                dist=norm(mat[i].reshape(1,-1)-mat[ind].reshape(1,-1))
                if dist<current_min:
                    current_min=dist
                    print(current_min)
                    min_ind=i
                    print(titles[min_ind])

    return min_ind
    
```

# Let's cluster them using k means

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(verbose=True,n_clusters=5,max_iter=300)
```

```python
kmeans.fit(t)
```

```python
x=[ (kmeans.predict(t[i].reshape(1,-1))[0],titles[i]) for i in range(len(t))]
```

```python
y=pd.DataFrame(x)
```

```python
y.columns=['cluster','title']
```
