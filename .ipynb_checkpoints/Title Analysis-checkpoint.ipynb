# In this, we'll look at frequently re-occuring words in video titles, and explore their statistics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data=pd.read_csv('data.csv')
titles_arr=np.array(data['title'])
```

```python
titles_str=' '.join(titles_arr)
```

```python
from collections import Counter
```

# we'll look at the most re-occuring titles, and assign them to groups

```python
Counter(titles_str.split(' ')).most_common()[:20]
```

# interesting groups are:
    1. بي_بي_سي_ترندينغ
    2. مصر
    3. سوريا
    4. حوار
    5. العالم

let's label these groups in the data frame

```python
data['group']=np.nan
```

```python
for word in ['عراق','سوري','العالم','حوار','مصر']:
    for i in range(len(data)):
        if word in data.iloc[i]['title']:
            data.iloc[i,data.columns.get_loc('group')]=word
```

```python
groups=data.groupby('group').sum()
```

```python
#normalizing, dividing by the number of videos in each group
groups=(groups.T/pd.Series(data['group'].value_counts().reindex(data.groupby('group').sum().index))).T
```

```python
data['group'].value_counts().reindex(data.groupby('group').sum().index).values
```

```python
groups
```

```python
groups['like_ratio']=groups['likes']/groups['dislikes']
```

```python
normalized=groups.copy()
```

```python
for col in groups.columns:
    normalized[col]/=groups[col].T.sum()
```

```python
normalized
```

```python
from bidi.algorithm import get_display
from bidi import algorithm as bidialg
import arabic_reshaper
from math import pi
def make_spider( df,row, title, color):
    categories=list(df)
    N = len(categories)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
    ax = plt.subplot(3,3,row+1, polar=True, )
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)
    plt.xticks(angles[:-1], categories, color='grey', size=8)
    ax.set_rlabel_position(0)
#     plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
    plt.ylim(0,0.4)
    values=df.iloc[row].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=2, linestyle='solid')
    ax.fill(angles, values, color=color, alpha=0.6)
    plt.title(title, size=11, color=color, y=1.1)


my_dpi=95
plt.figure(figsize=(1200/my_dpi, 900/my_dpi), dpi=my_dpi) 
# plt.tight_layout()
plt.subplots_adjust(hspace=0.55)
# Create a color palette:
my_palette = plt.cm.get_cmap("Set1", len(groups.index))
 
# Loop to plot
for row in range(0, 5):
    make_spider( normalized,row=row, title=get_display(arabic_reshaper.reshape(normalized.index[row])), color=my_palette(1))
```
